```python
# Force Colab to display graphics inline
%matplotlib inline

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# 1. Dataset
data = {
    'Region': ['Rural', 'Urban', 'Rural', 'Urban', 'Rural', 'Urban', 'Rural', 'Urban'],
    'Date': ['2019-12', '2019-12', '2020-04', '2020-04', '2020-08', '2020-08', '2021-12', '2021-12'],
    'Unemployment_Rate_Pct': [6.8, 8.5, 22.9, 25.1, 7.5, 9.8, 6.5, 7.9]
}
df = pd.DataFrame(data)
```

```python
# 2. Plotting everything in a single execution block
fig, ax = plt.subplots(figsize=(10, 5))
sns.lineplot(data=df, x='Date', y='Unemployment_Rate_Pct', hue='Region', marker='o', linewidth=2.5, ax=ax)

ax.set_title('Unemployment Rate Analysis (Pre-Covid vs Covid-19 Spike)', fontsize=14, fontweight='bold')
ax.set_xlabel('Timeline (Year-Month)', fontsize=12)
ax.set_ylabel('Unemployment Rate (%)', fontsize=12)
ax.grid(True, linestyle='--', alpha=0.6)
```

```python
# Explicitly call draw and show
plt.draw()
plt.show()
```
